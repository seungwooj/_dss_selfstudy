```python
np.random.seed(0)
df2 = pd.DataFrame({
    'key1' : ['A', 'A', 'B', 'B', 'A'],
    'key2' : ['one', 'two', 'one', 'two', 'one'],
    'data1' : [1,2,3,4,5],
    'data2' : [10,20,30,40,50]
})
df2
```

```python
df2.groupby(df2.key1).sum()
```

```python
### 특정열의 결과만 뽑기
a1 = df2.data1.groupby(df2.key1).sum()
a2 = df2.groupby(df2.key1)["data1"].sum()
a3 = df2.groupby(df2.key1).sum()["data1"]
print(a1, a2, a3)
```

```python
### 연습문제1 - key1기준으로 data1 값을 분류, 합계 구하고 DataFrame화
s1 = df2.groupby(df2.key1).sum()["data1"]
s1
```

```python
df3 = pd.DataFrame({"key1" : s1.index, "data1" : s1.values}) 
df3.sort_index(axis=1, ascending=False) ## data1, key1 순으로 정렬한것 순서 바꿔줌
```

```python
### 복합키 값에 따른 합계 구하기 - 피버세이블 형태로 만들기
df2["data1"].groupby([df2["key1"], df2["key2"]]).sum().unstack("key2")
```

```python
df1["인구"].groupby([df1["지역"], df1["연도"]]).sum().unstack("연도")
```

```python
### 붓꽃 case
import seaborn as sns
iris = sns.load_dataset("iris")
```

```python
iris.tail(5)
```

```python
### 붓꽃 종별 가장 큰값, 가장 작은값의 비율 -> agg함수로 함수 정의해서 사용
```

```python
def peak_to_peak_ratio(x):
    return x.max() / x.min()
```

```python
iris.groupby(iris["species"]).agg(peak_to_peak_ratio)
```

```python
### 붓꽃 데이터의 각종 기술통계 값 구하기 -> describe함수
iris.groupby(iris["species"]).describe().T
```

```python
### 붓꽃 데이터 안에서 종별로 가장 꽃잎길이가 큰 3개 데이터 뽑기 -> apply함수
def max3(df):
    return df.sort_values(by="petal_length")[:3]

iris.groupby(iris["species"]).apply(max3)
```

```python
### 붓꽃데이터 활용, "대/중/소"로 구분하기 -> transform함수

def q3cut(s):
    return pd.qcut(s, 3, labels = ["소", "중", "대"])

iris["petal_length_class"] = iris.groupby(iris.species)["petal_length"].transform(q3cut)
### <- petal_length_class라는 이름의 열 추가, petal_length기준으로 transform함수 적용
iris[["petal_length", "petal_length_class"]].tail(10) #<-아래 10개 데이터만 추출
```

```python
### 연습문제2 - species별 length, width 평균 나오는 df만들기
iris.tail(5)
```

```python
df_length = iris[["petal_length"]].groupby(iris["species"]).mean()
df_width = iris[["petal_width"]].groupby(iris["species"]).mean()
```

```python
df_length.join(df_width)
```

```python
### 예제 - Tip데이터 예제 - 식사대금 대비 팁의 비율이 가장 높아지는 경우 찾기
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips.tail()
```

```python
### 식사대금 대비 tip의 비중 열 추가
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips.tail()
```

```python
### 표 전반에 대한 분석
tips.describe()
```

```python
### 성별 데이터 갯수 세기
tips.groupby("sex").count()
```

```python
tips.groupby("sex").size()
```

```python
## 성별 + 흡연유무 - 데이터갯수
tips.pivot_table("tip_pct", "sex", "smoker", aggfunc="count", margins=True)
```

```python
#성별 + 흡연유무 - tip_pct의 평균
tips.pivot_table("tip_pct", "sex", "smoker")
```

```python
tips.groupby("sex")[["tip_pct"]].mean()
```

```python
tips.groupby("smoker")[["tip_pct"]].mean()
```

```python
### 성별과 흡연 - 기술통계값
tips.groupby(["sex", "smoker"])[["tip_pct"]].describe()
```

```python
tips.groupby("sex")[["tip_pct"]].describe()
```

```python
tips.groupby("smoker")[["tip_pct"]].describe()
```

```python
###1-1. 요일, 점심/저녁여부, 인원수 개별적으로 보기
```

```python
tips.groupby("day")["tip_pct"].mean()
```

```python
tips.groupby("time")["tip_pct"].mean()
```

```python
tips.groupby("size")["tip_pct"].mean()
```

```python
###1-2. 요일, 점심/저녁여부, 인원수 두개씩 보기
```

```python
tips.pivot_table("tip_pct", "day", "time", fill_value="")
```

```python
tips.pivot_table("tip_pct", "day", "size", fill_value="")
```

```python
tips.pivot_table("tip_pct", "time", "size", fill_value="")
```

```python
###1-3. 요일, 점심/저녁여부, 인원수 다 같이 보기
```

```python
tips.pivot_table("tip_pct", ["day", "time"], "size", fill_value="")
```

```python
tips.groupby(["day", "time", "size"])[["tip_pct"]].describe()
```

```python
### 성별/흡연여부별 팁평균, 가장 많은 팁, 가장 적은 팁의 차이 구하기 - agg활용

def peak_to_peak(x):
    return x.max() - x.min()

tips.groupby(["sex", "smoker"]).agg(["mean", peak_to_peak])[["total_bill"]] ###<- agg로 여러개 함수를 섞을 수 있다
```

```python
### 열마다 다른 연산을 할 수도 있음
tips.groupby(["sex", "smoker"]).agg({'tip_pct' : 'mean', 'total_bill' : peak_to_peak})
```

```python
### pivot_table활용한 좀 더 복잡한 연산

tips.pivot_table(['tip_pct', 'size'], ['sex', 'day'], 'smoker')
### tip_pct, size : value, sex, day : index, smoker : column
```
