```python
### 다중 인덱스
```

```python
np.random.seed(0)
df3 = pd.DataFrame(np.round(np.random.randn(6,4),2), 
                   columns=[["A","A","B","B"], ["C1","C2","C1","C2"]],
                  index=[["M","M","M","F","F","F"],
                        ["id_"+ str(i+1) for i in range(3)]*2])
df3.columns.names = ["Cidx1", "Cidx2"]
df3.index.names = ["Rid1", "Rid2"]
df3
```

```python
df3.stack("Cidx1")
```

```python
df3.stack(0)
```

```python
### 다중 인덱스의 인덱스 순서 교환
df4 = df3.swaplevel("Rid1", "Rid2")
df4
```

```python
df5 = df4.swaplevel("Cidx1", "Cidx2", axis=1)
df5
```

```python
### sort_index : 다중 인덱스의 정렬
```

```python
df5.sort_index(axis=1, level=0)
```

##### 06_데이터프레임의병합

```python
### 연습문제 1
```

```python
a = ["삼성","현대","SK", "롯데", "CJ"]
```

```python
np.random.seed(0)
df1 = pd.DataFrame(
    np.random.randint(10000,100000,size=(5,5)),
    columns = [2000,2001,2002,2003,2004])
df1
```

```python
df1.insert(loc=0,column="종목", value=a)
df1
```

```python
np.random.seed(1)
df2 = pd.DataFrame(
    np.random.randint(10000,100000,size=(5,5)),
    columns = [2005,2006,2007,2008,2009])
df2
```

```python
df2.insert(loc=0,column="회사", value=a)
df2
```

```python
pd.merge(df1,df2,left_on="종목", right_on="회사")
```

```python
### 연습문제 2
```

```python
data1 = {
    "1월" : [60, 20],
    "2월" : [70,40],
    "3월" : [80,50],
    "4월" : [90,35],
    "5월" : [100,50],
    "6월" : [80,35]
}
columns1 = ["1월", "2월", "3월", "4월", "5월", "6월"]
index1 = ["매출", "비용"]
df1 = pd.DataFrame(data1, index=index1, columns=columns1)
df1
```

```python
data2 = {
    "7월" : [60, 20],
    "8월" : [70,40],
    "9월" : [80,50],
    "10월" : [90,35],
    "11월" : [100,50],
    "12월" : [80,35]
}
columns2 = ["7월", "8월", "9월", "10월", "11월", "12월"]
index2 = ["매출", "비용"]
df2 = pd.DataFrame(data2, index=index2, columns=columns2)
df2
```

```python
df3=pd.concat([df1,df2],axis=1)
df3
```

```python
df3.loc["이익"] = df3.loc["매출"] - df3.loc["비용"]
df3
```

```python
df3["TTL"] = df3.sum(axis=1).astype(int)
df3
```
