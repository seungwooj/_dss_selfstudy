##### 4. WHERE : 검색조건 추가 (Filtering)

```python
# 나이가 30살 이상인 user 조회
mysql> SELECT *
    -> FROM user1
    -> WHERE age >= 30;
+---------+-------+----------------+------+------------+
| user_id | name  | email          | age  | rdate      |
+---------+-------+----------------+------+------------+
|       2 | peter | peter@daum.net |   33 | 2017-02-20 |
+---------+-------+----------------+------+------------+
1 row in set (0.00 sec)

# 등록일이 2018-01-01 이후인 user 조회
mysql> SELECT *
    -> FROM user1
    -> WHERE rdate >= "2018-01-01"
    -> ORDER BY user_id DESC;
+---------+-------+-----------------+------+------------+
| user_id | name  | email           | age  | rdate      |
+---------+-------+-----------------+------+------------+
|       4 | jin   | jin@naver.com   |   20 | 2018-03-31 |
|       3 | alice | alice@naver.com |   23 | 2018-01-05 |
|       1 | jin   | pdj@daum.net    |   20 | 2018-03-20 |
+---------+-------+-----------------+------+------------+
3 rows in set (0.00 sec)

# 등록일이 2017-01-01, 2018-03-01 사이인 user 조회 (BETWEEN사용)
mysql> SELECT *
    -> FROM user1
    -> WHERE rdate BETWEEN "2017-01-01" AND "2018-03-01";
+---------+-------+-----------------+------+------------+
| user_id | name  | email           | age  | rdate      |
+---------+-------+-----------------+------+------------+
|       2 | peter | peter@daum.net  |   33 | 2017-02-20 |
|       3 | alice | alice@naver.com |   23 | 2018-01-05 |
+---------+-------+-----------------+------+------------+
2 rows in set (0.00 sec)
```

##### 4. ORDER BY : sorting(정렬)

```python
# age로 오름차순
mysql> SELECT *
    -> FROM user1
    -> ORDER BY age ASC;
+---------+-------+-----------------+------+------------+
| user_id | name  | email           | age  | rdate      |
+---------+-------+-----------------+------+------------+
|       5 | andy  | andy@gmail.com  |   17 | 2016-04-28 |
|       1 | jin   | pdj@daum.net    |   20 | 2018-03-20 |
|       4 | jin   | jin@naver.com   |   20 | 2018-03-31 |
|       3 | alice | alice@naver.com |   23 | 2018-01-05 |
|       2 | peter | peter@daum.net  |   33 | 2017-02-20 |
+---------+-------+-----------------+------+------------+
5 rows in set (0.00 sec)

# age로 내림차순
mysql> SELECT *
    -> FROM user1
    -> ORDER BY age DESC;
+---------+-------+-----------------+------+------------+
| user_id | name  | email           | age  | rdate      |
+---------+-------+-----------------+------+------------+
|       2 | peter | peter@daum.net  |   33 | 2017-02-20 |
|       3 | alice | alice@naver.com |   23 | 2018-01-05 |
|       1 | jin   | pdj@daum.net    |   20 | 2018-03-20 |
|       4 | jin   | jin@naver.com   |   20 | 2018-03-31 |
|       5 | andy  | andy@gmail.com  |   17 | 2016-04-28 |
+---------+-------+-----------------+------+------------+
5 rows in set (0.00 sec)

# age로 내림차순, rdate로 오름차순 정렬
mysql> SELECT *
    -> FROM user1
    -> ORDER BY age DESC, rdate ASC;
+---------+-------+-----------------+------+------------+
| user_id | name  | email           | age  | rdate      |
+---------+-------+-----------------+------+------------+
|       2 | peter | peter@daum.net  |   33 | 2017-02-20 |
|       3 | alice | alice@naver.com |   23 | 2018-01-05 |
|       1 | jin   | pdj@daum.net    |   20 | 2018-03-20 |
|       4 | jin   | jin@naver.com   |   20 | 2018-03-31 |
|       5 | andy  | andy@gmail.com  |   17 | 2016-04-28 |
+---------+-------+-----------------+------+------------+
5 rows in set (0.00 sec)
```

##### 5. CONCAT : 합치기

```python
# name(age) 로 보여주기
mysql> SELECT email, CONCAT(name, "(", age, ")") AS "name(age)"
    -> FROM user1;
+-----------------+-----------+
| email           | name(age) |
+-----------------+-----------+
| pdj@daum.net    | jin(20)   |
| peter@daum.net  | peter(33) |
| alice@naver.com | alice(23) |
| andy@gmail.com  | andy(17)  |
| jin@naver.com   | jin(20)   |
+-----------------+-----------+
5 rows in set (0.00 sec)
```

##### 6. LIKE : 특정 문자열 들어간 데이터 찾기 (WHERE + LIKE 로 조회)

```python
# email 에 gmail 이 들어간 데이터 조회
mysql> SELECT *
    -> FROM user1
    -> WHERE email LIKE "%gmail%";
+---------+------+----------------+------+------------+
| user_id | name | email          | age  | rdate      |
+---------+------+----------------+------+------------+
|       5 | andy | andy@gmail.com |   17 | 2016-04-28 |
+---------+------+----------------+------+------------+
1 row in set (0.00 sec)

# email 에 gmail이 들어가지 않은 데이터 조회
mysql> SELECT *
    -> FROM user1
    -> WHERE email NOT LIKE "%gmail%";
+---------+-------+-----------------+------+------------+
| user_id | name  | email           | age  | rdate      |
+---------+-------+-----------------+------+------------+
|       1 | jin   | pdj@daum.net    |   20 | 2018-03-20 |
|       2 | peter | peter@daum.net  |   33 | 2017-02-20 |
|       3 | alice | alice@naver.com |   23 | 2018-01-05 |
|       4 | jin   | jin@naver.com   |   20 | 2018-03-31 |
+---------+-------+-----------------+------+------------+
4 rows in set (0.00 sec)
```

##### 6. IN : 여러 조건 만족하는 데이터 조회 : WHERE ~ IN ( 조건 )

```python
# name이 peter 와 alice
mysql> SELECT *
    -> FROM user1
    -> WHERE name IN ("peter", "alice");
+---------+-------+-----------------+------+------------+
| user_id | name  | email           | age  | rdate      |
+---------+-------+-----------------+------+------------+
|       2 | peter | peter@daum.net  |   33 | 2017-02-20 |
|       3 | alice | alice@naver.com |   23 | 2018-01-05 |
+---------+-------+-----------------+------+------------+
2 rows in set (0.00 sec)
```
