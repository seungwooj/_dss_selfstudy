```python
#4. 나라 전체 인구의 10%이상인 도시에서 도시인구가 500만이 넘는 도시를 아래와 같이 조회하세요.
mysql> SELECT city.Name, city.CountryCode, country.Name,
    -> ROUND(city.Population / country.Population * 100,2) AS percentage
    -> FROM city
    -> JOIN country
    -> ON city.CountryCode = country.Code
    -> HAVING percentage > 10
    -> AND city.Population > "5000000"
    -> ORDER BY percentage DESC;
+---------------------+-------------+----------------+----------+
| Name                | CountryCode | Name           |percentage|
+---------------------+-------------+----------------+----------+
| Lima                | PER         | Peru           |    25.19 |
| Seoul               | KOR         | South Korea    |    21.31 |
| Santafé de Bogotá   | COL         | Colombia       |    14.79 |
| Istanbul            | TUR         | Turkey         |    13.20 |
| London              | GBR         | United Kingdom |    12.22 |
| Bangkok             | THA         | Thailand       |    10.21 |
+---------------------+-------------+----------------+----------+
6 rows in set (0.01 sec)
```

```python
## 강사님 풀이

SELECT city.name, country.code, country.name,
ROUND(city.population / country.population * 100,2) AS percentage
FROM (SELECT * FROM city WHERE population > 5000000 )AS city
JOIN country
ON city.countrycode = country.code
HAVING percentage > 10
ORDER BY percentage DESC
```

```python
#5. 면적이 10000km^2 이상인 국가의 인구밀도(1km^2 당 인구수)를 구하고 
#인구밀도가 200이상인 국가들의 사용하고 있는 언어수가 5가지 이상인 나라를 조회 하세요.
mysql> SELECT name, language_count
    -> FROM
    -> (SELECT code, name,
    -> ROUND(Population / SurfaceArea) AS density
    -> FROM country
    -> WHERE SurfaceArea > 10000 AND
    -> ROUND(Population / SurfaceArea) > 200) AS country
    ->
    -> JOIN
    ->
    -> (SELECT CountryCode, count(Language) AS language_count
    -> FROM countrylanguage
    -> GROUP BY CountryCode
    -> HAVING count(Language) >= 5) AS countrylanguage
    ->
    -> ON country.Code = countrylanguage.CountryCode
    -> ORDER BY language_count DESC;
+-------------+----------------+
| name        | language_count |
+-------------+----------------+
| India       |             12 |
| Philippines |             10 |
| Vietnam     |              9 |
| Bangladesh  |              7 |
| Belgium     |              6 |
| Japan       |              6 |
| Germany     |              6 |
| Taiwan      |              6 |
+-------------+----------------+
8 rows in set (0.01 sec)
```

```python
## 강사님 풀이
SELECT country.name, count(country.name) AS language_count
FROM
(
SELECT code, name, ROUND(population / surfacearea) AS density
FROM country
WHERE surfacearea > 10000
HAVING density > 200
ORDER BY density DESC
) AS country
JOIN countrylanguage AS cl
ON cl.countrycode = country.code
GROUP BY name
HAVING language_count > 5
ORDER BY language_count DESC

## 코딩을 할때 자연어를 컴퓨터의 언어로 옮기는 연습에 익숙해지면 좋습니다.
```

```python
#6. 사용하는 언어가 3가지 이하인 국가중 도시인구가 300만 이상인 도시를 조회하세요.
# GROUP_CONCAT(LANGUAGE) 사용
# VIEW를 이용, query깔끔하게 수정

mysql> SELECT city_country.countrycode AS countrycode, city_name, population, name, language_count, languages
    ->
    -> FROM (
    ->
    -> SELECT city.countrycode AS countrycode, city_name, population, country.name
    -> FROM
    ->
    -> (SELECT CountryCode AS countrycode, Name AS city_name, Population AS population
    -> FROM city
    -> WHERE population > 3000000) AS city
    ->
    -> JOIN
    ->
    -> (SELECT Code, Name AS name
    -> FROM country) AS country
    ->
    -> ON city.CountryCode = country.Code
    ->
    -> ) AS city_country
    ->
    -> JOIN
    ->
    -> (SELECT CountryCode, COUNT(Language) AS language_count, GROUP_CONCAT(LANGUAGE) AS languages
    -> FROM countrylanguage
    -> GROUP BY CountryCode
    -> HAVING language_count <= 3) AS countrylanguage
    ->
    -> ON city_country.countrycode = countrylanguage.CountryCode
    -> ORDER BY population DESC;
+-------------+------------+------------+----------------+----------------+-------------------------+
| countrycode | city_name  | population | name           | language_count | languages               |
+-------------+------------+------------+----------------+----------------+-------------------------+
| KOR         | Seoul      |    9981619 | South Korea    |              2 | Chinese,Korean          |
| TUR         | Istanbul   |    8787958 | Turkey         |              3 | Arabic,Kurdish,Turkish  |
| GBR         | London     |    7285000 | United Kingdom |              3 | English,Gaeli,Kymri     |
| EGY         | Cairo      |    6789479 | Egypt          |              2 | Arabic,Sinaberberi      |
| PER         | Lima       |    6464693 | Peru           |              3 | Aimará,Ketšua,Spanish   |
| SGP         | Singapore  |    4017733 | Singapore      |              3 | Chinese,Malay,Tamil     |
| KOR         | Pusan      |    3804522 | South Korea    |              2 | Chinese,Korean          |
| EGY         | Alexandria |    3328196 | Egypt          |              2 | Arabic,Sinaberberi      |
| SAU         | Riyadh     |    3324000 | Saudi Arabia   |              1 | Arabic                  |
| TUR         | Ankara     |    3038159 | Turkey         |              3 | Arabic,Kurdish,Turkish  |
+-------------+------------+------------+----------------+----------------+-------------------------+
10 rows in set (0.01 sec)
```

```python
## 강사님 풀이
SELECT city.countrycode, city.name, city.population, country.name, city.language_count, city.languages
FROM 

(
SELECT city.countrycode, city.name, city.population, cl.language_count, cl.languages
FROM

(SELECT countrycode, group_concat(language) AS languages,
count(language) AS language_count
FROM countrylanguage
GROUP BY countrycode
HAVING language_count <= 3) AS cl

JOIN (SELECT * FROM city WHERE population > 3000000) AS city
ON city.countrycode = cl.countrycode
) AS city
JOIN country
ON city.countrycode = country.code
ORDER BY city.population DESC
```

```python
## VIEW 사용할 경우
CREATE VIEW cl AS

(
SELECT city.countrycode, city.name, city.population, cl.language_count, cl.languages
FROM

(
SELECT countrycode, group_concat(language) AS languages,
count(language) AS language_count
FROM countrylanguage

GROUP BY countrycode
HAVING language_count <= 3) AS cl

JOIN (SELECT * FROM city WHERE population > 3000000) AS city
ON city.countrycode = cl.countrycode
)

# 이후 쿼리문
SELECT city.countrycode, city.name, city.population, country.name,
city.language_count, city.languages
FROM cl AS city
JOIN country
ON city.countrycode = country.code
ORDER BY city.population DESC
```
