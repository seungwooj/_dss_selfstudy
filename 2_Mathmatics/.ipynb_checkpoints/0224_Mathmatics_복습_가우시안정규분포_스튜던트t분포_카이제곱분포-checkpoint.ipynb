```python
mu = 0
std = 1
rv = sp.stats.norm(mu,std)
```

##### 스튜던트 t분포

```python
### Fat tail 현상 예시 - 주식 수익률
```

```python
import pandas_datareader.data as web

symbols = ['SP500', 'NASDAQCOM', 'DJCA', 'NIKKEI225']
data = pd.DataFrame()
for sym in symbols:
    data[sym] = web.DataReader(sym, data_source='fred')[sym]
data = data.dropna()
(data / data.iloc[0] * 100).plot()

plt.figure(figsize=(10,8))
plt.show()
```

```python
log_returns = np.log(data / data.shift(1))
log_returns.hist(bins=50)
plt.show()
```

```python
for i, sym in enumerate(symbols):
    ax = plt.subplot(2, 2, i+1)
    sp.stats.probplot(log_returns[sym].dropna(), plot=ax)
plt.tight_layout()
plt.show()
```

```python
### 스튜던트 t분포
```

```python
xx = np.linspace(-4, 4, 100)
for df in [1, 2, 5, 10, 20]:
    rv = sp.stats.t(df=df)
    plt.plot(xx, rv.pdf(xx), label=("student-t (dof=%d)" %df))
plt.plot(xx, sp.stats.norm().pdf(xx), label="Normal", lw=5, alpha=0.5)
plt.legend()
plt.show()
```

```python
### 정규분포의 샘플평균이 이루는 분포
```

```python
np.random.seed(0)

rv = sp.stats.norm()
M = 1000

plt.subplot(1, 2, 1)
N = 4                       # <- N : 확률변수의 개수
x1 = rv.rvs((N, M))
xbar1 = x1.mean(axis=0)
xstd1 = x1.std(axis=0, ddof=1)
x = xbar1 / (xstd1 / np.sqrt(N))
sns.distplot(x, kde=True)
xx = np.linspace(-6, 6, 1000)
plt.plot(xx, rv.pdf(xx), 'r:', label="normal")
plt.xlim(-6, 6)
plt.ylim(0, 0.5)
plt.title("N = 4")
plt.legend()

plt.subplot(1, 2, 2)
N = 40
x2 = rv.rvs((N, M))
xbar2 = x2.mean(axis=0)
xstd2 = x2.std(axis=0, ddof=1)
x = xbar2 / (xstd2  / np.sqrt(N))
sns.distplot(x, kde=True)
xx = np.linspace(-6, 6, 1000)
plt.plot(xx, rv.pdf(xx), 'r:', label="normal")
plt.xlim(-6, 6)
plt.ylim(0, 0.5)
plt.title("N = 40")
plt.legend()

plt.show()
```

##### 카이제곱분포

```python
### 카이제곱분포 그래프
```

```python
xx = np.linspace(0.01, 10, 100)
for df in [1, 2, 3, 4, 5, 6, 9]:
    rv = sp.stats.chi2(df=df)
    plt.plot(xx, rv.pdf(xx), label=("chi2 (dof=%d)" %df))
plt.xlim(0, 10.1)
plt.ylim(0, 0.6)
plt.legend()
plt.show()
```

```python
np.random.seed(0)

rv = sp.stats.norm()
M = 1000

plt.subplot(1, 2, 1)
N = 1
x = rv.rvs((N, M))
t = (x**2).sum(axis=0)
sns.distplot(t, kde=False)
plt.xlim(-1, 10)
plt.ylim(0, 500)
plt.title("N=1")

plt.subplot(1, 2, 2)
N = 3
x = rv.rvs((N, M))
t = (x**2).sum(axis=0)
sns.distplot(t, kde=False)
plt.xlim(-1, 10)
plt.ylim(0, 500)
plt.title("N=3")

plt.show()
```

##### F분포

```python
xx = np.linspace(0.03, 3, 1000)
plt.plot(xx, sp.stats.f(1,1).pdf(xx), label="F(1,1)")
plt.plot(xx, sp.stats.f(5,1).pdf(xx), label="F(10,2)")
plt.plot(xx, sp.stats.f(10, 10).pdf(xx), label="F(20,20)")
plt.legend()
plt.show()
```

```python
np.random.seed(0)

rv = sp.stats.norm()
M = 10000

N1 = 10
x1 = rv.rvs((N1, M))     # <- rvs((N1, M)) : (N1,M) 사이즈의 행렬 random value로
t1 = (x1 ** 2).sum(axis=0)

N2 = 10
x2 = rv.rvs((N2, M))
t2 = (x2**2).sum(axis=0)

t = t2 / t1
sns.distplot(t, bins=200, kde=True)
plt.xlim(-0.1, 3)
plt.title("N1=10, N2=10")

plt.show()
```
