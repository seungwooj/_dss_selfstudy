# 1. 확률론적 생성모형
- "생성 모형" : 클래스를 지정하면 그에 맞는 값을 생성 할 수 있다. (모형을 알기 때문에) -> 새로운 데이터에 대한 값을 구할 수 있다.
- 베이즈 정리를 활용, Likelihood를 통해 조건부 확률을 구한다. <br>
$ P(y = C_k \mid x) \;\; \propto \;\; P(x \mid y = C_k) \; P(y = C_k) $
- 각 클래스 k에 대한 확률을 비교, 가장 확률이 큰 클래스 k를 선택한다.
- P(x)의 경우, 모든 P(y)를 구할 때 들어가므로 생략하기도 한다. <br>
<br>
Likelihood를 구하는 방법
    -   $ P(x \mid y = C_k) $  의 확률 분포모형을 가정
    - K번째 클래스의 트레이닝데이터 (독립변수) 활용하여 확률 분포 모형의 모수를 구한다. = $ P(x \mid y = C_k) $ 를 구한다.
    - 새로운 x에 대해 $ P(x \mid y = C_k) $ 계산하고 베이즈 정리를 활용하여 $ P(y = C_k \mid x) $ 를 구한다.

# 2. QDA
- 가장 단순한 확률론적 생성모형 : $ P(x \mid y = C_k) $  의 확률 분포모형을 "Gaussian Normal distribution"으로 가정
- $ p(x \mid y = k) = \dfrac{1}{(2\pi)^{D/2} |\Sigma_k|^{1/2}} \exp \left( -\dfrac{1}{2} (x-\mu_k)^T \Sigma_k^{-1} (x-\mu_k) \right) $
- $ P(y=k \mid x) = \dfrac{p(x \mid y = k)P(y=k)}{p(x)} = \dfrac{p(x \mid y = k)P(y=k)}{\sum_l p(x \mid y = l)P(y=l) } $
<br>
- 위의 식과 같이 독립변수 x에 대한 y의 조건부 확률을 1)베이즈정리, 2)전체확률의법칙 을 활용하여 구할 수 있다.
- QDA모형에서는 1)각각의 클래스에 대한 2) mu, 공분산 행렬을 구한다.


```python
# Scikit-Learn실습

N = 100
np.random.seed(0)

#sp.stats.multinomial(mean, cov)
X1 = sp.stats.multivariate_normal([0,0], [[0.7,0], [0,0.7]]).rvs(100)
X2 = sp.stats.multivariate_normal([1,1], [[0.8,0.2], [0.2,0.8]]).rvs(100)
X3 = sp.stats.multivariate_normal([-1,1], [[0.8,0.2], [0.2,0.8]]).rvs(100)

# 클래스가 0,1,2인 경우 생성
y1 = np.zeros(N)
y2 = np.ones(N)
y3 = 2 * np.ones(N)

X = np.vstack([X1, X2, X3])
y = np.hstack([y1, y2, y3])
```

```python
plt.scatter(X1[:,0], X1[:,1], alpha=0.8, s=50, color='r', label="class 1")
plt.scatter(X2[:,0], X2[:,1], alpha=0.8, s=50, color='g', label="class 2")
plt.scatter(X3[:,0], X3[:,1], alpha=0.8, s=50, color='b', label="class 3")
sns.kdeplot(X1[:,0], X1[:,1], alpha=0.3, cmap=mpl.cm.hot)
sns.kdeplot(X2[:,0], X2[:,1], alpha=0.3, cmap=mpl.cm.summer)
sns.kdeplot(X3[:,0], X3[:,1], alpha=0.3, cmap=mpl.cm.cool)
plt.xlim(-5,5)
plt.ylim(-4,5)
plt.legend()
plt.show()
```

```python
# QDA 수식 활용
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
qda = QuadraticDiscriminantAnalysis(store_covariance=True).fit(X,y)
```

```python
#QDA X1, X2의 각클래스에 대한 mu , covariance
print(qda.means_, '\n')
print(qda.covariance_[0], '\n')
print(qda.covariance_[1], '\n')
print(qda.covariance_[2])
```

```python
# QDA를 통해 예측한 결과 plotting

xmin, xmax = -5, 5
ymin, ymax = -4, 5
XX, YY = np.meshgrid(np.arange(xmin, xmax, (xmax-xmin)/1000), 
                     np.arange(ymin, ymax, (ymax-ymin)/1000))
ZZ = np.reshape(qda.predict(np.array([XX.ravel(), YY.ravel()]).T), 
                XX.shape)
cmap = mpl.colors.ListedColormap(sns.color_palette(["r", "g", "b"]).as_hex())
plt.contourf(XX, YY, ZZ, cmap=cmap, alpha=0.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap=cmap)
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)
plt.show()
```

# 3. LDA
- QDA에 가정을 1개 추가 : 클래스마다 mu는 다르지만, covariance matrix는 공통으로 가진다. <br>
- why? 
    1. Cov행렬이 많아질수록 (차원이 커질수록) 오차 증가.
    2. 현실적으로 클래스별로 mu는 차이가 커도 sigma는 크지 않을 경우가 많다.
    3. $\Sigma_k = \Sigma \;\;\; \text{ for all } k$ , 여러 데이터에서 sigma를 계산, 더 정확해진다.
<br>
- LDA의 결론 : $\Sigma_k = \Sigma \;\;\; \text{ for all } k$ 일때, $w^Tx + C = 0$, 즉 경계선은 직선이 된다.
- 경계선이 직선일때, LDA와 퍼셉트론의 결과가 동일하게 나온다.

```python
#Scikit-Learn실습
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=3, solver="svd", store_covariance=True).fit(X,y)
```

```python
print(lda.means_, '\n')
print(lda.covariance_)
```

```python
#LDA를 통해 예측한 내용 plotting -> 경계선이 직선임.

xmin, xmax = -5, 5
ymin, ymax = -4, 5
XX, YY = np.meshgrid(np.arange(xmin, xmax, (xmax-xmin)/1000), 
                     np.arange(ymin, ymax, (ymax-ymin)/1000))
ZZ = np.reshape(lda.predict(np.array([XX.ravel(), YY.ravel()]).T), XX.shape)
cmap = mpl.colors.ListedColormap(sns.color_palette(["r", "g", "b"]).as_hex())
plt.contourf(XX, YY, ZZ, cmap=cmap, alpha=0.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap=cmap)
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)
plt.show()
```
