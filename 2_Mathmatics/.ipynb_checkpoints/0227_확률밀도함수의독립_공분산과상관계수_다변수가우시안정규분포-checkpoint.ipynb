##### 확률밀도함수의 독립

```python
### 확률 밀도 함수의 독립
```

```python
np.set_printoptions(precision=4)
pmf1=np.array([[1, 2, 3, 2, 1],
              [2, 4, 6, 4, 2],
              [4, 8, 12, 8, 4],
              [2, 4, 6, 4, 2],
              [1, 2, 3, 2, 1]])
pmf1 = pmf1 / pmf1.sum()
pmf1
```

```python
sns.heatmap(pmf1, cmap=mpl.cm.bone_r)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Joint Probability (Independent)")
plt.show()
```

```python
pmf1_marginal_x = pmf1.sum(axis=0)
pmf1_marginal_y = pmf1.sum(axis=1)
pmf = pmf1_marginal_x * pmf1_marginal_y[:,np.newaxis]
pmf/pmf.sum()
```

```python
### y값이 뭐가 되도 조건부 확률은 동일
```

```python
cond_x_y0 = pmf1[0, :]/pmf1_marginal_y[0]
cond_x_y0
```

```python
cond_x_y1 = pmf1[1, :]/pmf1_marginal_y[1]
cond_x_y1
```

```python
### 독립 조건이 성립하지 않는 케이스
```

```python
pmf2 = np.array([[0, 0, 0, 1, 1],
                [0, 0, 2, 3, 1],
                [0, 2, 5, 2, 0],
                [1, 3, 2, 0, 0],
                [1, 1, 0, 0, 0]])
pmf2  = pmf2/pmf2.sum()
pmf2
```

```python
sns.heatmap(pmf2, cmap=mpl.cm.bone_r)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Joint Probability (Denpendent)")
plt.show()
```

```python
pmf2_marginal_x = pmf2.sum(axis=0)
pmf2_marginal_y = pmf2.sum(axis=1)
```

```python
### y값이 다를때 조건부 확률도 다름
```

```python
cond_x_y0 = pmf2[0, :]/pmf2_marginal_y[0]
cond_x_y0
```

```python
cond_x_y1 = pmf2[1, :]/pmf2_marginal_y[1]
cond_x_y1
```

##### 공분산과 상관계수

```python
np.random.seed(0)
corrs  = [1, 0.7, 0.3, 0., -0.3, -0.7, -1]
plt.figure(figsize=(len(corrs), 1))
for i, r in enumerate(corrs):
    x, y = np.random.multivariate_normal([0, 0], [[1, r], [r, 1]], 1000).T
    plt.subplot(1, len(corrs), i+1)
    plt.plot(x, y, 'ro', ms=1)
    plt.axis('equal')
    plt.xticks([])
    plt.yticks([])
    plt.title(r)
plt.show()
```

```python
np.random.seed(1)
slope =  [1, 0.7, 0.3, 0, 0.3, -0.3, -0.7, -1]
plt.figure(figsize=(len(slope), 1))
for i, s in enumerate(slope):
    plt.subplot(1, len(slope), i+1)
    x, y = np.random.multivariate_normal([0, 0], [[1, 1], [1, 1]], 1000).T 
    #<- mean = (0,0), cov xx = 1, cov xy = 1, cov yx = 1, cov yy = 1, 100개 찍기
    y2 = s * y
    plt.plot(x, y2, 'ro', ms=1)
    plt.axis('equal')
    plt.xticks([])
    plt.yticks([])
    if s > 0:
        plt.title(1)
    if s < 0:
        plt.title(-1)
plt.show()
```

##### 다변수 가우시안 정규분포

```python
### 경우 1
```

```python
mu = [2,3]
cov = [[1,0], [0,1]]
rv = sp.stats.multivariate_normal(mu, cov)
xx = np.linspace(0, 4, 120)
yy = np.linspace(1, 5, 150)
XX, YY = np.meshgrid(xx, yy)
plt.grid(False)
plt.contour(XX, YY, rv.pdf(np.dstack([XX, YY])))
plt.axis("equal")
plt.show()
```
