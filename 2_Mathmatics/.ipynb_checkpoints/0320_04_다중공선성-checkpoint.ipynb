```python
import pandas as pd
import numpy as np
import datetime as dt
import time
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
import statsmodels.api as sm # statsmodel 기본 import
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms 
import sklearn as sk
from patsy import dmatrix
```

##### 다중공선성 : 독립 변수의 일부가 다른 독립 변수의 조합으로 표현될 수 있는 경우 -> 상호상관관계가 강함, 공분산행렬이 full rank여야한다는 가정 침해

```python
from statsmodels.datasets.longley import load_pandas
dfy = load_pandas().endog
dfX0 = load_pandas().exog  # partial regression plotting
df = pd.concat([dfy, dfX0], axis=1)
sns.pairplot(dfX0)
plt.show()
```

```python
### 상관계수는 상관계수 행렬로도 살펴 볼 수 있음.
dfX0.corr()
```

```python
sns.heatmap(dfX0.corr())
plt.show()
```

```python
# 다중공선성이 있으면 독립변수간 상호 의존으로 과최적화 (over-fitting) 발생, 회귀 분석의 안정성 저하
# 독립변수의 공분산 행렬의 조건수 (cont'd no.) 증가, 가중치 행렬 계산시 오차 증가, 예측값의 오차도 증가
```

```python
# 해결1 : 스케일링 -> cont'd no. 감소 유도
```

```python
model = sm.OLS.from_formula("TOTEMP ~"
                           "scale(GNPDEFL) + scale(POP) + scale(GNP) +"
                           "scale(YEAR) + scale(ARMED) + scale(UNEMP)", data=df)
result = model.fit()
print(result.summary())
```

```python
# 해결2 : 변수 선택법 (VIF -> 의존적인 변수 발견, 제거 후 다시 OLS)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(dfX0.values, i) for i in range(dfX0.shape[1])]
vif["features"] = dfX0.columns
vif
```

```python
# VIF의 경우, 높을수록 해당 변수를 제외한 나머지로 회귀모델을 구성해도 모델의 성능이 좋음 . VIF값이 높은 변수들을 제외하고 모델을 돌려본다
```

```python
model2 = sm.OLS.from_formula("TOTEMP ~ scale(GNP) + scale(UNEMP) + scale(ARMED)", data=df)
result2 = model2.fit()
print(result2.summary())
```
