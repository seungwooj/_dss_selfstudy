# 모형 결합의 기초
- 모형결합(model combining) = 앙상블 방법론 (ensemble method)
: 복수의 예측 모형을 결합, 더 나은 성능 예측을 하려는 시도<br>
(개별 모형의 성능이 구려도 각각의 모형이 맞을 확률이 50% 만 넘는다면, 여러 모형을 합쳤을 때 성능이 향상된다!)<br>
<br>

<효과><br>
1. 단일 모형 대비 성능 분산의 감소 (과최적화 방지 : 여러 모형으로 동시에 판단하기 때문.)<br>
<-> 단일 의사결정 나무의 경우, 그 성능을 최대로 높일 수는 있으나, 과최적화의 문제가 존재한다.

2. 개별 모형의 성능이 50% 초과할 때, 개별 모형의 성능이 낮을수록 결합모형 성능의 개선 가능성은 더 증가함.<br>
(개별 모형에서 보완할 부분이 "많다" = 성능 개선의 여지가 많다.)<br>
<br>

<종류><br>
1. 취합방법론 (Aggregate method) : 사용할 모형의 집합이 이미 결정되어 있음. (동일한 목표의 모형 여러개)<br>
    = 열명이 똑같은 문제를 푼다 -> 미리 사용할 모형을 정해놔도 됨.<br>
    ex> 다수결(majority voting), 배깅(bagging), 랜덤포레스트(random forest)
<br>

2. 부스팅 방법론 (Boosting method) : 사용할 모형을 점진적으로 늘림 (앞선 모형의 부족한 점을 보완 - 목표가 서로 다른 모형 여러개) <br>
    = 앞사람이 못 푼 문제를 푼다 -> 미리 사용할 모형을 정할 수 없다.<br>
    ex> 에이다 부스트(Ada boost), 그레디언트 부스트(Gradient boost)


## 1. 다수결 방법  : 가장 단순, 전혀 다른 모형 결합 가능
- hard voting : 단순 투표 - 개별 모형의 결과 기준
- soft voting : 가중치 투표 - 개별 모형의 조건부 확률의 합 기준 -> 각 모형의 "비중"에 따라 클래스 결정을 달리 함.
    
[scikit-learn]<br>

sklearn.ensamble.VotingClassifier(estimators, voting='hard', weights=None)<br>

*입력인수
- estimators : 개별 모형 목록 - 리스트 or parameters
- voting : hard or soft (string)
- weights  : 가중치 리스트 (ex> [1,1,2])

```python
#예시용 plotting
X = np.array([[-1, -1], [-1.5, -1.5], [1, -0.5], 
              [-3.5, -2.5], [0, 1], [1, 1.5], [-2, -0.5]])
y = np.array([1, 1, 1, 2, 2, 2, 2])

plt.scatter(X[y==1,0], X[y==1,1], s=100, marker='o', c='r')
plt.scatter(X[y==2,0], X[y==2,1], s=100, marker='x', c='b')

x_new = [0.5, 0]

plt.scatter(x_new[0], x_new[1], s=100, marker='^', c='g')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.ensemble import VotingClassifier

#앙상블을 위한 개별 모델의 생성 및 앙상블 지정 (각각의 모델을 estimators로 받고, voting방법 및 weight지정)
model1 = LogisticRegression(random_state=1)
model2 = QuadraticDiscriminantAnalysis()
model3 = GaussianNB()
ensemble = VotingClassifier(estimators = [('lr', model1), ('qda', model2), ('gnb', model3)],
                         voting='soft', weights=[1,1,2])

#예측 확률을 comprehension을 활용, 행렬로 생성
probas = [c.fit(X,y).predict_proba([x_new]) for c in (model1, model2, model3, ensemble)]

class1_1 = [pr[0,0] for pr in probas]
class2_1 = [pr[0,1] for pr in probas]

#plotting
ind = np.arange(4)
width = 0.35  # bar width
p1 = plt.bar(ind, np.hstack(([class1_1[:-1], [0]])), width, color='green') #class_1에 대한 바차트
p2 = plt.bar(ind + width, np.hstack(([class2_1[:-1], [0]])), width, color='lightgreen') #class_2에 대한 바차트
p3 = plt.bar(ind, [0, 0, 0, class1_1[-1]], width, color='blue') #class_1에 대한 classifier바차트
p4 = plt.bar(ind + width, [0, 0, 0, class2_1[-1]], width, color='steelblue') #class_2에 대한 classifier바차트

plt.xticks(ind + 0.5 * width, ['LogisticRegression\nweight 1',
                               'QDA\nweight 1',
                               'GaussianNB\nweight 2',
                               'VotingClassifier'])
plt.ylim([0, 1.1])
plt.title('Class probabilities for sample 1 by different classifiers')
plt.legend([p1[0], p2[0]], ['class 1', 'class 2'], loc='upper left')
plt.show()
```

```python
from itertools import product

x_min, x_max = -4, 2
y_min, y_max = -3, 2
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.005), np.arange(y_min, y_max, 0.005))
f, axarr = plt.subplots(2, 2)
for idx, clf, tt in zip(product([0, 1], [0, 1]),
                        [model1, model2, model3, ensemble],
                        ['LogisticRegression', 'QDA', 'GaussianNB', 'VotingClassifier']):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    axarr[idx[0], idx[1]].contourf(xx, yy, Z, alpha=0.2, cmap=mpl.cm.jet)
    axarr[idx[0], idx[1]].scatter(X[:, 0], X[:, 1], c=y, alpha=0.5, s=50, cmap=mpl.cm.jet)
    axarr[idx[0], idx[1]].scatter(x_new[0], x_new[1], marker='x')
    axarr[idx[0], idx[1]].set_title(tt)
plt.tight_layout()    
plt.show()
```

### 모형 결합을 사용한 성능 향상 : 여러 모형으로 판단하는 것이 성능 자체가 좋다.<br>
(전제  : 1. 개별 모형의 정답확률 50% 이상 / 2. 서로 독립)
- 정답 출력 확률 : p, N개의 모형
- 다수결 모형의 정답 확률 : $\sum_{k>\frac{N}{2}}^N \binom N k p^k (1-p)^{N-k}$<br>
    ($k>\frac{N}{2}$ : 절반 이상이 맞아야 다수결 보형이 정답이 된다.)<br>
(시험 문제 혼자 푸는 것보다는 여러명이 같이 푸는게 더 낫다.)<br>
(단, 다같이 바보들이면 (50%미만) 무조건 틀린다.)


```python
def total_error(p, N):
    te = 0.0
    for k in range(int(np.ceil(N/2)), N + 1):
        te += sp.misc.comb(N, k) * p**k * (1-p)**(N-k)
    return te

x = np.linspace(0, 1, 100)
plt.plot(x, x, 'g:', lw=3, label="individual model")
plt.plot(x, total_error(x, 10), 'b-', label="voting model (N=10)")
plt.plot(x, total_error(x, 100), 'r-', label="voting model (N=100)")
plt.xlabel("performance of individual model")
plt.ylabel("performance of voting model")
plt.legend(loc=0)
plt.show()
```

## 2. 배깅 (bagging)
- 동일 모형, 동일 모수 사용하는 대신  "트레이닝 데이터를 랜덤하게 선택" <br>
=> 다수결 모형 적용 (트레이닝 데이터를 다르게 해서 여러번 테스트)

<종류>
- 같은 데이터 중복 사용 X : Pasting
- 같은 데이터 중복 사용 O : Bagging
- 독립 변수 중 일부만 선택 : Random subspaces
- 데이터도 일부, 독립변수도 일부만 선택 : Random patches
    
[scikit-learn] Bagging classifier
<인수>
- base_estimator : 기본 모형
- n_estimators  : 모형 갯수 (정수, 디폴트10)
- max_samples : 데이터 샘플 중 선택할 샘플의 수 혹은 비율 (정수 or 실수, 디폴트 1.0)
- max_features : 독립변수 중 선택할 차원의 수 or 비율 (정수 or 실수, 디폴트 1.0)
- bootstrap : 데이터 중복 사용 여부 (불리언, 디폴트 False)
- bootstrap_features : 차원 중복 사용 여부 (불리언, 디폴트 False)
- oob_score  : 성능 평가시 OOB(out-of-bag : 외부 데이터) 사용 여부 (불리언, 디폴트 False)

c.f> outlier의 경우, 전체 모형 중 일부만 받게 된다 (bagging진행 시) -> 다수결 모형 적용 시 그 영향력이 줄어드는 효과가 있음.

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import BaggingClassifier

iris = load_iris()
X, y = iris.data[:, [0,2]], iris.target

model1 = DecisionTreeClassifier().fit(X,y)
model2 = BaggingClassifier(DecisionTreeClassifier(), bootstrap_features=True, random_state=0).fit(X,y)

x_min, x_max = X[:, 0].min()-1, X[:,0].max() + 1
y_min, y_max = X[:, 1].min()-1, X[:,1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))

plt.figure(figsize=(8, 12))

#model1 : 의사결정나무 로 predict시 오버피팅 발생
plt.subplot(211)
Z1 = model1.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.contourf(xx, yy, Z1, alpha=0.6, cmap=mpl.cm.jet)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=1, s=50, cmap=mpl.cm.jet)

#model2 : 배깅 사용시 오버피팅 사라지는 효과
plt.subplot(212)
Z2 = model2.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
plt.contourf(xx, yy, Z2, alpha=0.6, cmap=mpl.cm.jet)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=1, s=50, cmap=mpl.cm.jet)

plt.tight_layout()
plt.show()
```

## 3. 랜덤 포레스트 (Random Forest)
: 의사결정모형을 개별 모형으로 사용하는 모형 결합방법<br>
    => Decision Tree의 개별 node마다 feature_selection을 새로 한다. (feature차원을 랜덤하게 줄이고 그안에서 확률 높은애들 고름.)<br>
    => 개별 노드 단위로 다른 feature를 기준으로 모형을 구성하므로, 개별 모형들간에 독립적일 가능성이 높다. 서로간의 상관관계 감소.<br>
    
c.f>1. 만약 노드마다 feature_selection을 다시 안할 경우, 부모 노드에 영향력 강한 feature가 있어서 그거로 분류를 해버리고,<br>
이후의 자식 노드들끼리는 어떻게 뽑건 그 결과가 비슷해진다.

<=> 랜덤 포레스트의 경우, 영향력이 큰 feature가 안 들어갈 수 있으므로 다른 feature들의 영향을 볼 수 있음.<br>
"복합적인 시각으로 분류가 가능"

* 바르셀로나라는 팀의 실력을 검증할 때, 계속 메시/수아레즈만 가지고 보면 팀 실력을 검증할 수 없음. (다른 선수들도 경기는 같이 뛰니까.)

c.f>2. Extremely Randomized Tree : 각 노드에서feature를 랜덤하게 선택. (주사위 던지기 : 모든 feature에 공평한 기회.)

```python
from sklearn import clone
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()

n_classes = 3 #클래스 개수
n_estimators = 30 #개별모형개수
plot_colors = "ryb"
cmap = plt.cm.RdYlBu
plot_step = 0.02
seed = 13

models = [DecisionTreeClassifier(max_depth=4),
          RandomForestClassifier(max_depth=4, n_estimators=n_estimators),
          ExtraTreesClassifier(max_depth=4, n_estimators=n_estimators)]

plot_idx = 1
plt.figure(figsize=(12, 12))
for pair in ([0, 1], [0, 2], [2, 3]): #0: 꽃받침 길이, 1: 꽃받침 폭, 2: 꽃잎 길이, 3: 꽃잎 폭
    for model in models:
        X = iris.data[:, pair]
        y = iris.target

        idx = np.arange(X.shape[0])
        np.random.seed(seed)
        np.random.shuffle(idx) #X,y에 idx를 랜덤하게 부여
        X = X[idx]
        y = y[idx]

        mean = X.mean(axis=0)
        std = X.std(axis=0)
        X = (X - mean) / std

        clf = clone(model)
        clf = model.fit(X, y)

        plt.subplot(3, 3, plot_idx)
        model_title = str(type(model)).split(".")[-1][:-2][:-len("Classifier")]
        if plot_idx <= len(models):
            plt.title(model_title)

        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                             np.arange(y_min, y_max, plot_step))

        if isinstance(model, DecisionTreeClassifier):
            Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
            Z = Z.reshape(xx.shape)
            plt.contourf(xx, yy, Z, cmap=cmap)
        else:
            estimator_alpha = 1.0 / len(model.estimators_)
            for tree in model.estimators_:
                Z = tree.predict(np.c_[xx.ravel(), yy.ravel()])
                Z = Z.reshape(xx.shape)
                plt.contourf(xx, yy, Z, alpha=estimator_alpha, cmap=cmap)

        for i, c in zip(range(n_classes), plot_colors):
            idx = np.where(y == i)
            plt.scatter(X[idx, 0], X[idx, 1], c=c,
                        label=iris.target_names[i], cmap=cmap)

        plot_idx += 1

plt.tight_layout()
plt.show()
```

* 랜덤 포레스트 모델의 장점 : feature importance를 측정할 수 있다.<br>
(모든 노드에 대해 특정 feature로 사용시의 I.G를 구할 수 있음)<br>
-> 각각의 노드에서 특정 feature를 사용했을 때의 I.G평균값으로 어떤 독립변수가 중요한지 알 수 있음. (높을수록 중요)


* 특히 extremely randomized tree의 경우, 모든 feature을 사용할 확률이 동일하기 때문에 feature importance를 좀 더 분명하게 알 수 있음.

```python
from sklearn.datasets import make_classification
from sklearn.ensemble import ExtraTreesClassifier

X, y = make_classification(n_samples=1000, n_features=10, n_informative=3, n_redundant=0, n_repeated=0,
                           n_classes=2, random_state=0, shuffle=False) #독립적인 변수는 일부러 3개만 넣음.

forest = ExtraTreesClassifier(n_estimators=250, random_state=0)
forest.fit(X, y)

importances = forest.feature_importances_

std = np.std([tree.feature_importances_ for tree in forest.estimators_], axis=0)
indices = np.argsort(importances)[::-1]

print("Feature ranking:")
for f in range(X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))

plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices], color="r", yerr=std[indices], align="center")
plt.xticks(range(X.shape[1]), indices)
plt.xlim([-1, X.shape[1]])
plt.show()
```
