```python
def f(x, y):
    return 2 * x**2 + 6 * x * y + 7 * y**2 - 26 * x - 54 * y + 107
```

##### surface plot 그리기

```python
xx = np.linspace(-5, 5, 100)
yy = np.linspace(-4, 4, 100)
X, Y = np.meshgrid(xx, yy)
Z = f(X, Y)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
fig.gca(projection='3d').plot_surface(X, Y, Z)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Surface plot")
plt.show()
```

##### contour plot 그리기

```python
CS = plt.contour(X, Y, Z, levels=np.arange(0, 300, 20))
plt.clabel(CS, fmt="%d")
plt.title('Contour plot')
plt.show()
```

##### 3차원 softmax함수 연습

```python
def softmax(x, w):          # x, w의 가중합. x sms 1차원 배열
    e = np.exp(w * x)
    return np.exp(w * x) / e.sum()

x = [2, 1, 0.5]
y = softmax(x, np.ones(3))
y
```

```python
softmax(x, 4 * np.ones(3))     #<-가중치가 커지면 최대값, 최소값 차이가 더 벌어짐.
```
