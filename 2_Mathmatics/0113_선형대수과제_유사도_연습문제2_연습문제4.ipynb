### 1. 유사도 - 문제



```python
import numpy as np

import scipy as sp
import sklearn as sk

import matplotlib as mpl
mpl.use('Agg')
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
%matplotlib inline

from sklearn.datasets import load_digits

digits = load_digits()
d0 = digits.images[0]; d1 = digits.images[10]
d2 = digits.images[1]; d3 = digits.images[11]
v0 = d0.reshape(64,1)
v1 = d1.reshape(64,1)
v2 = d2.reshape(64,1)
v3 = d3.reshape(64,1)

f, ax = plt.subplots(4, 1, figsize=(12, 2))
ax[0].imshow(v0.T, interpolation='nearest', cmap=plt.cm.bone_r)
ax[0].grid(False); ax[0].set_xticks([]); ax[0].set_yticks([])
ax[1].imshow(v1.T, interpolation='nearest', cmap=plt.cm.bone_r)
ax[1].grid(False); ax[1].set_xticks([]); ax[1].set_yticks([])
ax[2].imshow(v2.T, interpolation='nearest', cmap=plt.cm.bone_r)
ax[2].grid(False); ax[2].set_xticks([]);ax[2].set_yticks([])
ax[3].imshow(v3.T, interpolation='nearest', cmap=plt.cm.bone_r)
ax[3].grid(False); ax[3].set_xticks([]);ax[3].set_yticks([])
plt.tight_layout()
plt.show()
```

```python
## 연습문제 2
# MNIST digit image 전체 데이터를 모두 벡터로 변환


# 8:8 행렬을 64:1 벡터로 전환한거 1797개 -> 64 X 1797 행렬 X 구하기
print(len(digits.images))
```

## 연습문제 2-(1)

```python
new_mx = []
for i in digits.images:
    reshaped = i.reshape(64, 1)
    new_mx.append(reshaped.T)

print(np.array(new_mx).T)
```

## 연습문제 2 - (2) 
- 2-(1)에서 구한 행렬을 'final'이라고 할 때,
  final의 전치행렬과 final의 곱은 모든 벡터들의 내적값을 원소로 가지는
  Matrix 가 된다. (정방행렬)

## 연습문제 4 - (1)

```python
final = np.array(new_mx).T.squeeze()  # <--- shape에서 1 삭제

np.matmul(final.T,final)    # <----matmul : 행렬의 곱 method
```

## 연습문제 4 - (2)

```python

answer = np.matmul(final.T, final)
imgplot = plt.imshow(answer, interpolation='nearest', cmap=plt.cm.bone_r)
plt.colorbar()
plt.show()
```
